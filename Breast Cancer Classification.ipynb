```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('C:/Users/tamil/Downloads/data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
data=df.dropna(axis=1)
```

```python
data= data.drop("id",axis=1)
```

```python
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data.drop(columns='diagnosis', axis=1)
Y = data['diagnosis']
X
```

```python
Y
```

```python
data.groupby('diagnosis').mean()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
print(X.shape, X_train.shape, X_test.shape)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, Y_train)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, Y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
  
knn.fit(X_train, Y_train)
```

```python
#for logical regression
# accuracy on training data
X_train_prediction = logmodel.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
#for logical regression
# accuracy on test data
X_test_prediction = logmodel.predict(X_test)
atest_log = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data = ', atest_log)
```
